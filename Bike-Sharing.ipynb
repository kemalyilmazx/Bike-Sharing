```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bikeshare= pd.read_csv("day.csv")
```

```python
## If day is Saturday or Sunday, working day is 0; else 1
bikeshare.head()
```

```python
bikeshare.info()
```

```python
bikeshare.describe()
```

```python
## If it is 2011, yr column is ; else 1 
bikeshare["year_2012"]= pd.get_dummies(bikeshare["yr"], drop_first= True)
```

```python
##
bikeshare.head(10)
```

```python
## Linear model would include square of temperature
bikeshare["squared_temp"]= bikeshare["temp"]**2
```

```python
## If weathersit is 1, then the weather is clear
bikeshare["clearday"]= np.where(bikeshare["weathersit"] == 1, 1, 0)
```

```python
import statsmodels.formula.api as smf
```

```python
bikeshare_df= bikeshare[["registered","temp", "squared_temp", "workingday", "clearday", "year_2012"]]
```

```python
## Creating a fitted model
lm1 = smf.ols(formula='registered ~ temp+squared_temp+workingday+clearday+year_2012', data=bikeshare_df).fit()
```

```python
# Coefficients
coefficients = pd.DataFrame(lm1.params)
coefficients.columns = ["Coefficients"]
coefficients
```

```python
## 95% Confidence Interval
CI = pd.DataFrame(lm1.conf_int())
CI.columns = ["Lower Confidence Interval", "Upper Confidence Interval"]
CI
```
